```python
import pandas as pd
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
x = pd.DataFrame(iris["data"], columns = iris["feature_names"])
x
```

```python
y = pd.DataFrame(iris["target"], columns = ["target_names"])
y
```

```python
data = pd.concat([x,y], axis = 1)
data
```

Call out another array "target" in iris dataset and use "target_name" as column names

#Using concat( ) to combine iris dataset and target array.

```python
data = pd.concat([x,y], axis = 1)
data
```

```python
from sklearn import preprocessing as pp
from sklearn import impute as ip
#print(data.mean())
#scale()是對數據進行z-score標準化
scaled = pp.scale(data)
pp.normalize(data)
pp.binarize(scaled,threshold=-1) #使用z-score後再套入二分法/threshold預設為0
```

#Use DataFrame type to do binarize
applymap(lambda ) x> -1 --> 1 else 0\
*applymap speculatly use DataFrame type

```python
df_scaled = pd.DataFrame(scaled, columns = data.columns)
print(df_scaled.head(3))
df_scaled.applymap(lambda x:1  if x>-1 else 0)
transformer = pp.FunctionTransformer(lambda scaled:[1 if y > -1 else 0 for y in x]for x in scaled)
print(transformer)
```

#Use transformer( ) to do binary preprocessing

```python

scaled = pp.scale(df_scaled)
print(scaled[:3])
transformer = pp.FunctionTransformer(lambda x: np.array([[1 if y > -1 else 0 for y in row] for row in x]))
transformed_data = transformer.transform(scaled)
print(transformed_data[:3])
```
